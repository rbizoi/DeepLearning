```python
import numpy as np, pandas as pd, seaborn as sns, warnings, os, sys, pickle, time
from matplotlib import pyplot as plt
from datetime import datetime as dt

import matplotlib.font_manager as fm
from matplotlib.lines import Line2D

warnings.filterwarnings(action="ignore")

if int(str(sns.__version__).split('.')[1]) > 8 : 
    plt.style.use('seaborn-v0_8-darkgrid')
else:
    plt.style.use('seaborn-darkgrid')
    
sns.set(font_scale=2)
```

```python
def repeat_tanh(x, n):
    v = x
    for i in range(0, n):
        v = np.tanh(v)
    return v
```

```python
xs = np.linspace(-8, 8, 500)
sns.set(font_scale=2)
plt.figure(figsize=(18,18))
plt.plot(xs, repeat_tanh(xs, 1   ), lw=2, c='black' , linestyle='-', label="1 tanh")
plt.plot(xs, repeat_tanh(xs, 2   ), lw=2, c='black' , linestyle='--', label="2 tanh") 
plt.plot(xs, repeat_tanh(xs, 5   ), lw=2, c='black' , linestyle='-.', label="5 tanh")
plt.plot(xs, repeat_tanh(xs, 25  ), lw=2, c='black' , linestyle=':', label="25 tanh")
plt.plot(xs, repeat_tanh(xs, 50  ), lw=2, c='blue'  , linestyle='--', label="50 tanh")
plt.plot(xs, repeat_tanh(xs, 100 ), lw=2, c='green' , linestyle='--', label="100 tanh")
plt.plot(xs, repeat_tanh(xs, 200 ), lw=2, c='yellow', linestyle='--', label="200 tanh")
plt.plot(xs, repeat_tanh(xs, 2000), lw=2, c='black' , linestyle='--', label="2000 tanh")
plt.plot(xs, repeat_tanh(xs, 4000), lw=2, c='red'   , linestyle='--', label="4000 tanh")
plt.legend(loc='best')
plt.xlabel("input")
plt.ylabel("output")
plt.title("Repeated tanh")
plt.show()
```
